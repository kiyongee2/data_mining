```python
# 연령별 인구수 시각화
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
result = []   # 연령별 인구 저장 리스트

for row in data:
    if '신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))  # 정수로 변환
            
#print("0~100세 개수 :", len(row[3:]))
print(result)
plt.style.use("ggplot")  # 격자 무늬 스타일 지정
plt.plot(result)
plt.show()
```

```python
# 지역을 입력받아 꺽은선 그래프 그리기

f = open('age.csv')
data = csv.reader(f)
result = []   # 연령별 인구 저장 리스트
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요:")

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))  # 정수로 변환   
            
plt.title(name + " 지역의 인구 구조")
plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot")  # 격자 무늬 스타일 지정
plt.plot(result)
plt.show()
```

```python
# 막대 그래프 그리기

f = open('age.csv')
data = csv.reader(f)
result = []   # 연령별 인구 저장 리스트
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요:")

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))  # 정수로 변환   
            
plt.title(name + " 지역의 인구 구조")
plt.rc('font', family='Malgun Gothic')
plt.bar(range(101), result)
plt.show()
```
