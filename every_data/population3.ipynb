### 인구 구조 시각화하기

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

f = open('Data/age2.csv')
data = csv.reader(f)
next(data)
home = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')

for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int)
print(home)
plt.rc('font', family='Malgun Gothic')
plt.title(name + " 지역의 인구 구조")
plt.plot(home)
plt.show()
```

```python
f = open('Data/age2.csv')
data = csv.reader(f)
next(data)
# data = list(data)
home = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')

for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int)/int(row[2])
#print(home)
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title(name + " 지역의 인구 구조")
plt.plot(home)
plt.show()
```

```python
f = open('Data/age2.csv')
data = csv.reader(f)
next(data)
home = []
away = []

for row in data:
    if "신도림" in row[0]:
        home = np.array(row[3:], dtype=int)/int(row[2])
    if "화촌면" in row[0]:
        away = np.array(row[3:], dtype=int)/int(row[2])

plt.rc('font', family='Malgun Gothic')
plt.title("신도림동 지역과 불광제1동 지역의 인구 구조 비교")
plt.plot(home)
plt.plot(away)
plt.show()
```

```python
f = open('Data/age2.csv')
data = csv.reader(f)
next(data)
data = list(data)  # data를 list로 저장
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')
min_v = 1
result_name = ""
result = []   # 최솟값을 갖는 지역의 연령대별 인구 비율을 저장할 배열 생성 및 초기화

for row in data:
    if name in row[0]:
        home = np.array(row[3:], dtype=int)/int(row[2])
for row in data:
    away = np.array(row[3:], dtype=int) / int(row[2])
    #print(home - away)
    #print(np.sum(home-away))
    sum_v = np.sum((home-away)**2)
    if sum_v < min_v and name not in row[0]:
        min_v = sum_v
        result_name = row[0]
        result = away

plt.rc('font', family='Malgun Gothic')
plt.title(name + " 지역과 가장 비슷한 인구 구조를 가진 지역")
plt.plot(home, label=name)
plt.plot(result, label=result_name)
plt.legend()
plt.show()
```
