```python
# 생일의 최고 기온 시각화 하기
import csv

f = open("seoul.csv")
data = csv.reader(f)
next(data)
result = [] # 최고 기온 데이터 저장

for row in data:
    row[0] = row[0].split('\t')[1]
    #print(row[0].split('-')[0])
    if row[-1] != "":
        if row[0].split('-')[1] == "08" and row[0].split('-')[2] == "15":
            result.append(float(row[-1]))
#print(result)
plt.plot(result, 'hotpink')
```

```python
# 1983년 이후 2월 14일 최고 기온 데이터 시각화

f = open("seoul.csv")
data = csv.reader(f)
next(data)
high = [] # 최고 기온 데이터 저장
low = []  # 최저 기온 데이터 저장

for row in data:
    row[0] = row[0].split('\t')[1]
    if row[-1] != "" and row[-2] !="":
        if 1983 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == "02" and row[0].split('-')[2] == "14":
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title("내 생일의 기온 변화 그래프")
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend()
plt.show()
```
