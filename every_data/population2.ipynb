## 서울 불광제1동 지역의 남여 성별 인구 분포

### 불광제1동 남성과 여성 인구수 시각화

```python
import csv

f = open('Data/gender.csv')
data = csv.reader(f)

header = next(data)
#print(header)

#print(data)

for row in data:
    if '신림동' in row[0]:
        print(row[1], row[105])  # 남성 총인구, 여성 총인구
```

```python
import matplotlib.pyplot as plt

f = open('Data/gender.csv')
data = csv.reader(f)

m = []
fm = []

for row in data:
    if '신림동' in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for i in row[106:]:
            fm.append(int(i))

#print(m)
#print(len(m), len(fm))

plt.plot(m, 'red')    # 꺽은선 그래프
#plt.bar(range(101), m)   # 막대 그래프
plt.plot(fm)
plt.show()
```

### 항아리 모양의 그래프 그리기

```python
f = open('Data/gender.csv')
data = csv.reader(f)

m = []
fm = []
name = input("찾고 싶은 지역명을 입력하세요 : ")

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i.replace(',','')))
        for i in row[106:]:
            fm.append(int(i.replace(',','')))
        break

#print(m, fm)
print(row[1], row[105])

plt.style.use('ggplot')
plt.figure(figsize=(8, 4), dpi=100)
plt.rc('font', family='Malgun Gothic')
plt.title(name + " 지역의 남녀 성별 인구 분포 ")
#plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m, label='남성')
plt.barh(range(101), fm, label='여성')
plt.legend()
plt.show()
```

### 제주도의 성별 인구 비율 표현하기

```python
import csv
import matplotlib.pyplot as plt

with open('Data/gender.csv') as f:
    data = csv.reader(f)

    size = []
    name = input("찾고 싶은 지역명을 입력하세요 : ")

    for row in data:
        if name in row[0]:
            sum_m = 0     # 남성 합계
            sum_fm = 0    # 여성 합계

            for i in range(101):
                sum_m += int(row[i+3].replace(',', ''))
                sum_fm += int(row[i+106].replace(',', ''))
            break
        
size.append(sum_m)
size.append(sum_fm)
print(size)

plt.title(name + " 지역의 남녀 성별 비율")
color = ['crimson', 'darkcyan']
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color,
       startangle=90)
plt.show()
```
