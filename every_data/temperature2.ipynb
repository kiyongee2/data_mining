## 내 생일의 최고기온과 최저기온 시각화

```python
import matplotlib.pyplot as plt
import csv

# 1983년 이후 2월 14일 최고 기온 데이터 시각화

with open("seoul.csv") as f:
    data = csv.reader(f)
    next(data)
    high = [] # 최고 기온 데이터 저장
    low = []  # 최저 기온 데이터 저장

    for row in data:
        row[0] = row[0].split('\t')[1]
        if row[-1] != "" and row[-2] !="":
            if 1993 <= int(row[0].split('-')[0]):
                if row[0].split('-')[1] == "02" and row[0].split('-')[2] == "14":
                    high.append(float(row[-1]))
                    low.append(float(row[-2]))
print(len(low))
print(high)
print(low)
plt.rc('font', family='Malgun Gothic')       # 한글 글꼴 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스(-)가 깨질때
plt.title("내 생일의 기온 변화 그래프")
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.xlabel("연도")
plt.ylabel("기온")
plt.legend()
plt.show()
```
