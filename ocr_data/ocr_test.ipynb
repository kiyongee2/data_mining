```python
from PIL import Image
import pytesseract
```

```python
pytesseract.pytesseract.tesseract_cmd = "C:/Program Files/Tesseract-OCR/tesseract.exe"
```

```python
img_file = Image.open("news.png")
img_file
```

```python
text = pytesseract.image_to_string(img_file, lang='kor')
#print(text)
print(text.replace(" ", ""))
```

```python
with open('news2.txt', 'w', encoding="utf8") as f:
    #f.write(text)
    f.write("\n\n")
    f.write(text.replace(" ", ""))
```
