## 데이터 시각화 
### matplotlib 라이브러리 사용
### 기본그래프 - plot() 함수 - 직선 또는 꺽은선 형태의 그래프

```python
import matplotlib.pyplot as plt

# 리스트는 y축값이며 x축값은 0부터 1씩 증가하는 정수로 입력됨
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
plt.rc("font", family="Malgun Gothic") #한글 글꼴 지원
plt.title("꺽은선 그래프")  # 제목

# 첫번째 리스트는 x축값, 두번째 리스트는 y축값임
plt.plot([1, 2, 3, 4], [12, 43, 25, 15])
plt.show()
```

```python
plt.title("범례")
plt.plot([10, 20, 30, 40], label='증가')
plt.plot([40, 30, 20, 10], label='감소', color='pink')
plt.legend()
plt.show()
```

### bar()는 막대그래프
### 막대의 길이는 각 데이터의 크기를 의미함

```python
plt.bar([1, 2, 3, 4], [12, 43, 25, 15])
plt.show()
```

```python
plt.rc("font", family="Malgun Gothic") #한글 글꼴 지원
plt.title("막대 그래프")  # 제목

plt.bar(range(6), [1, 2, 3, 5, 6, 7]) # range(0, 5)
plt.show()
```
