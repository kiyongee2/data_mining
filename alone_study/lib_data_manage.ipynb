```python
import pandas as pd

# 일괄 처리 함수
def data_cleaning(filename):
    ns_df = pd.read_csv(filename, low_memory=False)
    ns_book = ns_df.dropna(axis=1)   # NaN 열 삭제
    # 대출 건수를 합치기 위해 필요한 행만 추출
    count_df = ns_book[['도서명', '저자', 'ISBN','대출건수']]
    group_df = count_df.groupby(by=['도서명', '저자', 'ISBN'], dropna=False)
    loan_count = group_df.sum()
    dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN']) # 중복된 행을 True로 만듬
    unique_rows = ~dup_rows  #반전 : 중복되지 않은 행을 True로 만듬
    ns_book2 = ns_book[unique_rows].copy()
    return ns_book2
```

```python
new_ns_book = data_cleaning('./data/namsan_book2.csv')
new_ns_book.head(100)
```
