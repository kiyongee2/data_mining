### 불필요한 데이터 삭제하기

### 열 삭제하기

```python
selected_columns = ns_df.columns !='부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head(3)
```

```python
ns_book = ns_df.drop('부가기호', axis=1)
ns_book.head(3)
```

```python
# inplace=True는 현재 데이터프레임을 수정함
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

```python
# NaN이 하나 이상 포함된 행이나 열을 삭제
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

### 행 삭제하기

```python
# 행 삭제하기 - axis=0이 생략됨
ns_book2 = ns_book.drop([0, 1])
ns_book2.head()
```

```python
# 행 선택하기
ns_book2 = ns_book[2:]
ns_book2.head()
```

```python
selected_rows = ns_df['출판사'] == '한빛미디어'
ns_book2 = ns_book[selected_rows]
ns_book2.head()
```

```python
ns_book2 = ns_book[ns_book['대출건수'] >= 5]
ns_book2.head()
```

```python
# 중복된 행 찾기 - 중복되면 True를 1로 인식함
#sum(ns_book.duplicated())
sum(ns_book.duplicated(subset=['도서명', '저자', 'ISBN']))
```

```python
du_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)
ns_book3 = ns_book[du_rows]
ns_book3.head()
```

```python
# 그룹별로 모으기
count_df = ns_book[['도서명', '저자', 'ISBN', '대출건수']]
group_df = count_df.groupby(by=['도서명', '저자', 'ISBN'], dropna=False)
loan_count = group_df.sum()
loan_count.head()
```
